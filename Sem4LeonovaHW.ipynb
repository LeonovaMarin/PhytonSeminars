```python
import numpy as np
```

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm');
```

```python
sns.jointplot(x=df['bedrooms'], y=df['price'], kind='reg');
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['view'], y=df['price'],  whis=1.5)

plt.ylabel('Стоимость');
plt.xlabel('Вид из окна')
plt.title('Зависимость цены дома от вида из окна');
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['bedrooms'], y=df['price'],  whis=4)

plt.ylabel('Стоимость');
plt.xlabel('Количество спален')
plt.title('Зависимость цены дома от количества спален');
```
