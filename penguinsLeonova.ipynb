```python
penguins.loc[penguins.bill_length_mm >= 42, 'bill_group'] = 'High'
penguins.loc[(penguins.bill_length_mm >35) & (penguins.bill_length_mm < 42), 'bill_group'] = 'Middle'
penguins.loc[penguins.bill_length_mm <= 42, 'bill_group'] = 'Low'
penguins.head()
```

```python
penguins.fillna(penguins.mean())
```

```python
penguins[['bill_length_mm', 'bill_len']]
penguins.groupby('bill_len')['bill_length_mm'].mean()
penguins.groupby('bill_len')['bill_length_mm'].mean().plot(kind='bar')
```

```python
penguins.loc[penguins.flipper_length_mm >= 213, 'height_group'] = 'High'
penguins.loc[(penguins.flipper_length_mm >190) & (penguins.flipper_length_mm < 213), 'height_group'] = 'Middle'
penguins.loc[penguins.flipper_length_mm <= 190, 'height_group'] = 'Low'
penguins.head()
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='height_group')
```
