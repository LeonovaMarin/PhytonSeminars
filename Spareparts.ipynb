```python
df.drop(columns=['Изделие(child)', 'Unique key'], inplace=True)
df.head(2)
#можем удалить из таблицу ненужные столбцы (там описания позиций дублируются, но иногда они остаются пустыми.
#Можно оставить для поиска один наиболее подробный столбец, но я не определяла какой самый подходящий.
#Можно сделать это на более позднем этапе.
```

```python
#создаем словарь di, в котором указываем список для замены кодов заводов на текст: '661' - 'Vologda' и т.д. Т.е. коду 661 соответствует завод в Вологде, 662 = в Волгограде и пр.

di = {661: 'Vologda', 662: 'Volgograd', 663: 'Perm', 664: 'OmskLSD', 665: 'Angarsk', 666: 'Kazan', 668: 'Ryazan', 669: 'OmskCTR', 670: 'Beloyarsk', 672: 'KalugaLSD', 747: 'KalugaIBC'}
```

```python
df['Компания'] = df['Компания'].replace(di)
#при помощи функции replace меняем коды заводов на текст, которые мы записали в словарь di.
#Меняем значения только в столбце "Компания"
```

```python
df.head(3)
#убеждаемся, что в столбце "компания" коды переименовались в заводы.
```

```python
df = df.replace('>', '', regex=True)
df.head(3)
#при помощи функции replace удаляем лишний знак '>', заменяем его на пусто ''. Выводим первые 3 строчки для проверки.
```

```python
df.fillna('', inplace= True )
df.head(3)
#при помощи функции fillna заменяем пустые ячейки NaN на пустое строковое значение '' (иначе функция не работает).
#Проверяем, что получилось (вместо NaN - пусто) при помощи df.head. В столбце Quantity остаются числа 0.0 - значение не пустое, его можно посчитать.
```

```python
df = df[df.Quantity != 0.0]
df.head(3)
#добавляем условие, что в колонке "Quantity" (количество) убираем все нулевые (0.0) значения (Quantity != 0.0)
#df.head - Проверяем что получилось, теперь первая строчка не 0, а 3.
```

```python
#создаем список discard, в который вносим ненужные позиции (поиск по словам "лак", "перчатки", "лист", "одежда", "краска") из определенного столбца ["Коммерческое описание"]
discard = ["лак", "перчатки", "лист", "одежда", "краска", "костюм", "монитор", "Lenovo", "DAREX", "очки", "нарукавники", "футболка", "эмаль", "сандали", "бочка", "\bкрас"]

# создаем новую функцию, new_df, в которой удалены строки согласно списку discard по столбцу "Коммерческое описание"
new_df= df[~df["Коммерческое описание"].str.contains('|'.join(discard))]
```

```python
new_df.describe()
#описание очищенной от 0-ых значений и ненужных позиций таблицы. Осталось 1921 значение (из 9014). Минимальное значение 1.
```

```python
print(new_df['Компания'].value_counts())
#подсчет наличия запчастей по заводам (по столбцу "компания").
```

```python
search = new_df["Коммерческое описание"].str.contains(r'\bпневмо')
#Если нужно искать в определенном столце слово с неполным соответствием - например "пневмо". Ставим в поиск в виде r'\bпневмо'.
#создаем переменную search
```

```python
#выбираем в  столбце "коммерческое описание" строки, которые содержат определенное слово "модуль" (полное соответствие слова в поискt)
#search = df["Коммерческое описание"].str.contains('модуль')
res = new_df.loc[search]
#создаем переменную res для вывода результатов поиска
```
